```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
data=pd.read_csv('bestsellers.csv')
```



                                      EXPLORING THE DATASET
                                      
                                      

```python
# Displaying first 5 rows
data.head(5)
```

```python
# no.of rows and colums in data
data.shape
```

```python
# column names of the data
print(data.columns)
```

```python
# summary of the data
data.describe()
```

```python
# data types in the columns
data.dtypes
```

```python
#checking for the null values
data.isna().sum()
```

```python
empty_priced=(data['Price']==0).sum()
print(empty_priced)
```



                                       MODIFYING THE DATASET
                                       
                                       

```python
data.drop_duplicates(inplace=True)
data
```

```python
# checking if duplicates are removed
data.duplicated().sum()
```

```python
# renaming columns
data.rename(columns={'Name':'Title','Year':'Publication Year'},inplace=True)
data.head(5)
```

```python
# changing the data type
data['Price'].astype(float)
```

```python
# replacing the price=0 columns with mean
data.loc[data['Price']==0,'Price']=data.loc[data['Price']!=0,'Price'].mean()
(data['Price']==0).sum()
```



                                        SORTING THE DATASET
                                        


```python
sorted_by_Author=data.sort_values(by=['Author'])
sorted_by_Author
```

```python
#sorting the data set by year in ascending and user rating in descending
data.sort_values(by=['Publication Year','User Rating'],ascending=[True,False])
```

```python
#resetting the index
data.reset_index(inplace=True,drop=True)
data
```



                                          AGGREGATING THE DATASET
                                          
                                      

```python
# most frequently occuring author
data.Author.mode()
```

```python
data.Reviews.median()
```

```python
grouped_data=data.groupby(['Author'])
grouped_data.first()
```

```python
# group by
avg_rating=data.groupby(['Genre'])['User Rating'].mean()
print(avg_rating)
```

```python
# group by max
avg_rating.max()
```

```python
author_counts=data.Author.value_counts()
author_counts
```

```python
# Using pivot table
pd.pivot_table(data,index='Genre',values='User Rating',aggfunc='mean')
```

```python
pd.pivot_table(data,index=['Title'],values='User Rating',aggfunc='count')
```


    
                                       EXPORTING THE RESULTS
                                       
                                    

```python
data.to_csv('bestsellers_modified.csv')
```

```python
author_counts.head(10).to_csv('top_authors.csv')
```

```python
avg_rating.to_csv('avg_rating_by_genre.csv')
```



                                        VISUALIZING THE DATASET
                                        
                        

```python
plt.figure(figsize=(10,5))
author_counts.head(10).plot(kind='bar',color='darkblue')
plt.title('Top 10 Authors')
plt.xlabel('Author Name')
plt.ylabel('No.of books')
plt.xticks(rotation=90)
plt.show()
```

```python
genre_count=data['Genre'].value_counts()
genre_count.plot(kind='pie',autopct='%1.1f%%',colors=['lightgreen', 'orange'])
plt.title('Distribution of Genres')
plt.ylabel('')
plt.show()
```

```python
sn.catplot(x='Price',kind='violin',data=data,color='yellow')
plt.title('Distribution of Book Prices')
plt.show()
```
